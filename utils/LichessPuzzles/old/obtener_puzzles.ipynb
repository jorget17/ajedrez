```python
import pandas as pd
import numpy as np
import time
import chess
from datetime import datetime
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
def genera_posicion_problema(inicial, moves):
  tablero = chess.Board(inicial)
  movimiento = chess.Move.from_uci(moves.split()[0])
  tablero.push(movimiento)
  problema = tablero.fen()
  return problema
```

```python
pregunta = input('Descargo el fichero? (Pulsa s para descargarlo): ')

if pregunta.lower() == "s":
  !wget https://database.lichess.org/lichess_db_puzzle.csv.bz2
  !bzip2 -d lichess_db_puzzle.csv.bz2
else:
  pass
```

```python
l = pd.read_csv('lichess_db_puzzle.csv', names=['puzzle_id', 'fen', 'moves', 'rating', 'deviation', 'popularity', 'played', 'themes', 'game'])
```

```python
inicio = time.perf_counter()

l['problema'] = l.apply(lambda x: genera_posicion_problema(x['fen'], x['moves']), axis=1)
l['sp_san'] = l.apply(lambda x: chess.Board(x['problema']).variation_san([chess.Move.from_uci(m) for m in x['moves'].split()[1:]]), axis=1)

final = time.perf_counter()

duracion = final -inicio

print(f'Proceso terminado en {duracion} segundos.')
```

```python
with open('lichess_puzzles.csv', 'w') as output_csv:
  for index, row in l.iterrows():
    output_csv.write(f"{row['problema']}|{row['puzzle_id']}|{row['game']}|{row['rating']}|{row['sp_san']}\n")
```

```python
with open('lichess_puzzles.pgn', 'w') as output_pgn:
  for index, row in l.iterrows():
    puzzle_id = row['puzzle_id']
    rating = row['rating']
    game = row['game']
    fen = row['problema']
    solucion = row['sp_san']
    played = row['played']
    output_pgn.write(f'''[White "Lichess problem {puzzle_id}"]
[Black "{rating}"]
[Event "{played}"]
[Site "{game}"]
[FEN "{fen}"]

{solucion} *
\n''')
```
